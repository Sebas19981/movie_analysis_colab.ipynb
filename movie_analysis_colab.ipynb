Este cuaderno realiza un an치lisis de datos b치sicos usando Python y pandas sobre el dataset `movie_metadata.csv`. Incluye limpieza, estad칤sticas descriptivas, an치lisis espec칤ficos y visualizaci칩n.

```python
# Paso 1: Carga de librer칤as y datos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# Cargar dataset (reemplazar con ruta si se descarga localmente)
df = pd.read_csv("movie_metadata.csv")
df.head()
```

```python
# Informaci칩n general
df.info()
```

```python
# Estad칤sticas descriptivas
df.describe()
```

## Paso 2: Limpieza de datos y nulos

```python
# Ver cu치ntos valores faltan en 'duration'
df['duration'].isnull().sum()
```

```python
# Reemplazar NaNs con la mediana de la duraci칩n
mediana_duration = df['duration'].median()
df['duration'] = df['duration'].fillna(mediana_duration)
```

```python
# Duraci칩n promedio redondeada
promedio = round(df['duration'].mean(), 2)
print(f"Duraci칩n promedio: {promedio} minutos")
```

## Paso 3: An치lisis espec칤fico

```python
# Asegurar tipo num칠rico
df['title_year'] = pd.to_numeric(df['title_year'], errors='coerce')

# Filtrar pel칤culas del 2008 con duraci칩n entre 90 y 120
peliculas_2008 = df[(df['title_year'] == 2008) & (df['duration'] >= 90) & (df['duration'] <= 120)]
print(f"N칰mero de pel칤culas lanzadas en 2008 con duraci칩n entre 90 y 120 min: {len(peliculas_2008)}")
```

```python
# Limpiar y convertir 'budget'
df['budget'] = df['budget'].replace('[\$,]', '', regex=True)
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')

# Mediana del presupuesto
mediana_budget = int(df['budget'].median())
print(f"Mediana del presupuesto: {mediana_budget}")
```

## Paso 4: An치lisis adicional

```python
# Gr치fico de barras por rango de duraci칩n vs presupuesto promedio
df['rango_duracion'] = pd.cut(df['duration'], bins=[0, 60, 90, 120, 150, 180, 300])
promedio_por_rango = df.groupby('rango_duracion')['budget'].mean().dropna()

plt.figure(figsize=(10,6))
sns.barplot(x=promedio_por_rango.index.astype(str), y=promedio_por_rango.values, palette='coolwarm')
plt.title('Presupuesto promedio por rango de duraci칩n')
plt.xlabel('Rango de duraci칩n (min)')
plt.ylabel('Presupuesto promedio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Detectar outliers en duraci칩n y graficar con barras
q1 = df['duration'].quantile(0.25)
q3 = df['duration'].quantile(0.75)
iqr = q3 - q1
lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr

df['es_outlier'] = df['duration'].apply(lambda x: 'Outlier' if (x < lim_inf or x > lim_sup) else 'Normal')
conteo = df['es_outlier'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=conteo.index, y=conteo.values, palette='viridis')
plt.title('Pel칤culas normales vs. outliers por duraci칩n')
plt.ylabel('Cantidad de pel칤culas')
plt.xlabel('Tipo de duraci칩n')
plt.show()
```
